### Fine_Grained Selection Words

```python
vocab = set(text1)
long_words = [w for w in vocab if len(w) > 10]
sorted(long_words)
```

```python
vocab2 = set(text2)
short_words = [word for word in vocab2 if len(word) == 5]
print(short_words)
```

```python
vocab3 = set(text1)
ends_ly = [w for w in vocab3 if w.endswith('ly') and len(w) >= 12]
print(ends_ly)
```

```python
fdist = FreqDist(text1)
sorted(w for w in set(text1) if len(w) == 12 and fdist[w] > 3)
```

### looping With Conditions

```python
sent_1 = ['text', 'one', 'is', 'here', '.']
for w in sent_1:
    if w.startswith('t'):
        print(w)
```

```python
for token in sent2:
    if token.islower():
        print(token, 'is a lowercase word')
    elif token.istitle():
        print(token, 'is a titlecase word')
    else:
        print(token, 'is punctuation')
```
